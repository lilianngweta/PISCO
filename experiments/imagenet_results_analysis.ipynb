```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import json, os, glob, re

import matplotlib
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['text.latex.preamble'] = [
    r'\usepackage{amsmath}',
    r'\usepackage{amssymb}']

from matplotlib import rc
rc("text", usetex=False)

import seaborn as sns
sns.set_style("whitegrid")
```

```python
path = './results/results_imagenet/'

list_dir = []

for file in glob.glob(path + '*.json'):
    with open(file) as f:
        data = json.load(f)
        list_dir.append(data)
        
df = pd.DataFrame(list_dir)
# print(df.columns)
# df
```

# Top 1 accuracy

```python
keys = ['dataset','extractor','transformation_type', 'epochs', 'batch_size','learning_rate', 'eta', 'lamda']
measure = ['baseline_indist_acc_no_sp_corr', 'PISCO_indist_acc_no_sp_corr', 'baseline_ood_acc_no_sp_corr','PISCO_ood_acc_no_sp_corr']

variables = {}
for index in measure:
    variables[index] = ['mean']
# summary = df.groupby(keys, as_index=False).agg(variables)
summary1 = df.groupby(keys, as_index=False).agg(variables)
summary1 = summary1[summary1.eta==0.95]
summary1 = summary1[summary1.batch_size==32768] #8192,32768
summary1 = summary1[summary1.learning_rate==0.0001]
summary1 = summary1[summary1.epochs==50]

summary1_lamda1 = summary1[summary1.lamda==1]
summary1_lamda10 = summary1[summary1.lamda==10]
summary1_lamda50 = summary1[summary1.lamda==50]


# summary1
```

```python
# Generating table for latex for top-1 accuracies 
# ***************
baseline_ood_no_sp_corr_mean = summary1_lamda1.baseline_ood_acc_no_sp_corr["mean"]
lamda1_PISCO_ood_no_sp_corr_mean = summary1_lamda1.PISCO_ood_acc_no_sp_corr["mean"]
lamda10_PISCO_ood_no_sp_corr_mean = summary1_lamda10.PISCO_ood_acc_no_sp_corr["mean"]
lamda50_PISCO_ood_no_sp_corr_mean = summary1_lamda50.PISCO_ood_acc_no_sp_corr["mean"]

baseline_indist_acc_lamda1 = list(summary1_lamda1.baseline_indist_acc_no_sp_corr["mean"].apply("{:.3f}".format))[0]
pisco_indist_acc_lamda1 = list(summary1_lamda1.PISCO_indist_acc_no_sp_corr["mean"].apply("{:.3f}".format))[0]
pisco_indist_acc_lamda10 = list(summary1_lamda10.PISCO_indist_acc_no_sp_corr["mean"].apply("{:.3f}".format))[0]
pisco_indist_acc_lamda50 = list(summary1_lamda50.PISCO_indist_acc_no_sp_corr["mean"].apply("{:.3f}".format))[0]


columns = ["Transformation"]
table1 = pd.DataFrame(list(summary1_lamda1.transformation_type)+["none"], columns=columns)
table1['Baseline (ResNet 50)'] = list(baseline_ood_no_sp_corr_mean)+ [baseline_indist_acc_lamda1]
# table1['Baseline (ResNet 50)'] = list(baseline_ood_no_sp_corr_mean.apply("{:.3f}".format)+ '$\pm$' + 
#                                      baseline_ood_no_sp_corr_std.apply("{:.3f}".format))
table1['PISCO,'+r' $\lambda = 1$'] = list(lamda1_PISCO_ood_no_sp_corr_mean)+ [pisco_indist_acc_lamda1]
table1['PISCO,'+r' $\lambda = 10$'] = list(lamda10_PISCO_ood_no_sp_corr_mean)+ [pisco_indist_acc_lamda10]
table1['PISCO,'+r' $\lambda = 50$'] = list(lamda50_PISCO_ood_no_sp_corr_mean)+ [pisco_indist_acc_lamda50]

display(table1)

# print(table1.to_latex(escape=False, column_format='l' + 'c'*len(table1.columns)))
```

```python
#  Generating table for latex for top-5 accuracies   
# ***************
baseline_ood_no_sp_corr_mean_t5 = summary2_lamda1.baseline_ood_t5_acc_no_sp_corr["mean"]
lamda1_PISCO_ood_no_sp_corr_mean_t5 = summary2_lamda1.PISCO_ood_t5_acc_no_sp_corr["mean"]
lamda10_PISCO_ood_no_sp_corr_mean_t5 = summary2_lamda10.PISCO_ood_t5_acc_no_sp_corr["mean"]
lamda50_PISCO_ood_no_sp_corr_mean_t5 = summary2_lamda50.PISCO_ood_t5_acc_no_sp_corr["mean"]

baseline_indist_t5_acc_lamda1 = list(summary2_lamda1.baseline_indist_t5_acc_no_sp_corr["mean"].apply("{:.3f}".format))[0]
pisco_indist_t5_acc_lamda1 = list(summary2_lamda1.PISCO_indist_t5_acc_no_sp_corr["mean"].apply("{:.3f}".format))[0]
pisco_indist_t5_acc_lamda10 = list(summary2_lamda10.PISCO_indist_t5_acc_no_sp_corr["mean"].apply("{:.3f}".format))[0]
pisco_indist_t5_acc_lamda50 = list(summary2_lamda50.PISCO_indist_t5_acc_no_sp_corr["mean"].apply("{:.3f}".format))[0]

columns = ["Transformation"]
table2 = pd.DataFrame(list(summary2_lamda1.transformation_type)+["none"], columns=columns)
table2['Baseline (ResNet 50)'] = list(baseline_ood_no_sp_corr_mean_t5)+[baseline_indist_t5_acc_lamda1]
# table1['Baseline (ResNet 18)'] = list(z_ood_no_sp_corr_mean.apply("{:.3f}".format)+ '$\pm$' + 
#                                      z_ood_no_sp_corr_std.apply("{:.3f}".format))
table2['PISCO,'+r' $\lambda = 1$'] = list(lamda1_PISCO_ood_no_sp_corr_mean_t5)+[pisco_indist_t5_acc_lamda1]
table2['PISCO,'+r' $\lambda = 10$'] = list(lamda10_PISCO_ood_no_sp_corr_mean_t5)+[pisco_indist_t5_acc_lamda10]
table2['PISCO,'+r' $\lambda = 50$'] = list(lamda50_PISCO_ood_no_sp_corr_mean_t5)+[pisco_indist_t5_acc_lamda50]

display(table2)

# print(table2.to_latex(escape=False, column_format='l' + 'c'*len(table1.columns)))
```

# Final combined table

```python
# add_top_column function from https://stackoverflow.com/questions/51021468/can-sub-columns-be-created-in-a-pandas-data-frame
def add_top_column(df, top_col, inplace=False):
    if not inplace:
        df = df.copy()
    
    df.columns = pd.MultiIndex.from_product([[top_col], df.columns])
    return df


columns = ["Transformation"]
df1 = pd.DataFrame(list(summary1_lamda1.transformation_type)+["none"], columns=columns)

# Baseline (ResNet 50)
df2 = pd.DataFrame(list(baseline_ood_no_sp_corr_mean.apply("{:.3f}".format))+[baseline_indist_acc_lamda1], columns=['Top-1'])
df2['Top-5'] = list(baseline_ood_no_sp_corr_mean_t5.apply("{:.3f}".format))+[baseline_indist_t5_acc_lamda1]
df2 = add_top_column(df2, 'Baseline (ResNet 50)')

# PISCO (lambda = 1)
df3 = pd.DataFrame(list(lamda1_PISCO_ood_no_sp_corr_mean.apply("{:.3f}".format))+[pisco_indist_acc_lamda1], columns=['Top-1'])
df3['Top-5'] = list(lamda1_PISCO_ood_no_sp_corr_mean_t5.apply("{:.3f}".format))+[pisco_indist_t5_acc_lamda1]
df3 = add_top_column(df3, 'PISCO,'+r' $\lambda = 1$')

# PISCO (lambda = 10)
df4 = pd.DataFrame(list(lamda10_PISCO_ood_no_sp_corr_mean.apply("{:.3f}".format))+[pisco_indist_acc_lamda10], columns=['Top-1'])
df4['Top-5'] = list(lamda10_PISCO_ood_no_sp_corr_mean_t5.apply("{:.3f}".format))+[pisco_indist_t5_acc_lamda10]
df4 = add_top_column(df4, 'PISCO,'+r' $\lambda = 10$')

# PISCO (lambda = 50)
df5 = pd.DataFrame(list(lamda50_PISCO_ood_no_sp_corr_mean.apply("{:.3f}".format))+[pisco_indist_acc_lamda50], columns=['Top-1'])
df5['Top-5'] = list(lamda50_PISCO_ood_no_sp_corr_mean_t5.apply("{:.3f}".format))+[pisco_indist_t5_acc_lamda50]
df5 = add_top_column(df5, 'PISCO,'+r' $\lambda = 50$')


table3 = pd.concat([df2, df3, df4, df5], axis=1)

table3["Style"] = list(summary1_lamda1.transformation_type)+["none"]

# shift column 'Style' to first position
first_column = table3.pop('Style')
  
# insert column using insert(position,column_name,first_column) function
table3.insert(0, 'Style', first_column)


display(table3)

# print latex code for the generated table
print(table3.to_latex(escape=False, column_format='l' + 'c'*len(table3.columns)))
```
